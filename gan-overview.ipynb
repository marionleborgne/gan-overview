## Introduction

* GANs = Generative Adversarial Networks 
* A relatively new type of generative models
* Generative models are a branch of unsupervised learning in ML

## The intuitive idea behind GANs




* Imagine an aspiring painter who wants to do **art forgery (G)**, and someone who wants to earn his living by **judging paintings (D)**. 


* You start by showing D some examples of work by Picasso. 


* Then G produces paintings in an attempt to fool D every time, making him believe they are Picasso originals. 


* Sometimes G succeeds. However as D starts learning more about Picasso style (looking at more examples), G has a harder time fooling D, so he has to do better. 


* As this process continues, not only D gets really good in telling apart what is Picasso and what is not, but also G gets really good at forging Picasso paintings. 



This is the idea behind GANs :-)


## How GANs really work

3 notions to understand about GANs


* Generative model (G)
    * In charge of determining the undelying distribution of the data

* Discriminative model (D)
    * In charge of determining the decision boundary between classes in the data

* Adversarial game
    * The two models D and G play an adversarial game.
    * What is an adversarial game? 
        * A game where player1's goal is to increase some value X and player2's goal is to decrease X

## Training GANs

The training process is in 2 steps



* Step1: Only the discriminator D participates
    * D is trained on sample data
    * D should output a probability close to 1


* Step 2: Both the discriminator D and the generator G participate
    * D is trained on fake data generated by G
    * G generates a fake image based on random noise.
    * D should output a probability close to 0

## Other considerations for GAN training


The D network is **simply a binary classifier trained with supervised learning**


* Half should be classified as 1 and the other as 0


* But **as a whole, the GAN system is unsupervised** because D's training data is labelled automatically by G
    * 1/2 training examples for D are real samples from the data (labelled ast 1)
    * 1/2 training examples for D are fake samples generated by G (labelles as 0)


The two networks **G and D must not have the same optimization function**


* We need to use a different gradient to backpropagate through D 
    * Because D plays against G. 


* A simple way to do that is **take the negative value network G's gradient** and **continue to backpropagate it through D**.
    * D's goal: tell wether the data is real or fake
    * G's goal: produce data to fool the discriminator

## Vector space arithmetic



* Code vectors
    * Random noise to specify which image should be generated by the generator


* Take several random noise values that all decoded to a same class of image and average them
    * This gives you the code for the average image representing this class


* You can do arithmetic on these codes
    * Decode several variations of a class code vector to get several variation of images in this class
    * Similar to word embeddings like word2vec or cortical.io
        * Harder because the output needs to make sense
        * The resulting word embedding does not need to make sense but just to be close to a couple of other words
